```python
# Dependencies
import requests
import json
import pandas as pd

# Import the API key
from config import eia_key
```

```python
#Input URL for the state and year (plus your API code)
full_url = f"https://api.eia.gov/v2/electricity/facility-fuel/data/?frequency=annual&data[0]=gross-generation&facets[state][]=CA&facets[fuel2002][]=BIT&facets[fuel2002][]=NG&facets[fuel2002][]=NUC&facets[fuel2002][]=RC&facets[fuel2002][]=SUN&facets[fuel2002][]=WAT&facets[fuel2002][]=WND&start=2021&end=2022&sort[0][column]=period&sort[0][direction]=desc&offset=0&length=5000&{eia_key}"
```

```python
#Request info from EIA API
CA_data2022 = requests.get(full_url).json()
```

```python
#View results
print(json.dumps(CA_data2022, indent=4, sort_keys=True))
```

```python
#A set of empty lists to store the information
plant_names = []
generation_amounts_mwh = []
fueltypes = []
years = []
states = []
```

```python
#Go through the data and add the information to each list
for i in range(len(CA_data2022["response"]["data"])):
    plant_names.append(CA_data2022["response"]["data"][i]["plantName"])
    generation_amounts_mwh.append(CA_data2022["response"]["data"][i]["gross-generation"])
    fueltypes.append(CA_data2022["response"]["data"][i]["fuelTypeDescription"])
    years.append(CA_data2022["response"]["data"][i]["period"])
    states.append(CA_data2022["response"]["data"][i]["state"])
print(len(plant_names))
print(len(generation_amounts_mwh))
print(len(fueltypes))
print(len(years))
print(len(states))
```

```python
CA_2022_df = pd.DataFrame({"Plant Name" :plant_names,
                           "Generation Amount (MWh)":generation_amounts_mwh,
                          "Fuel Type" : fueltypes,
                          "Year": years,
                          "State": states})
CA_2022_df
```

```python
CA_2022_df.to_csv("Output/CA_2022.csv")
```
